```sql
CREATE CATALOG IF NOT EXISTS dbacademy;
CREATE SCHEMA IF NOT EXISTS dbacademy.ops;
DROP TABLE IF EXISTS dbacademy.ops.meta_source;
```

```sql
%python

spark.sql('''
  CREATE TABLE IF NOT EXISTS dbacademy.ops.meta_source (
    owner VARCHAR(255) NOT NULL,
    object VARCHAR(255),
    key VARCHAR(255),
    value VARCHAR(255)
)
''')

def insert_into_meta_source(user_email:str, 
                            user_name_no_email:str, 
                            catalog_name: str, 
                            schema_name: str):
  spark.sql(f'''
    INSERT INTO dbacademy.ops.meta_source (owner, object, key, value) VALUES
    ("{user_email}", NULL, "username", "{user_email}"),
    ("{user_email}", NULL, "catalog_name", "{catalog_name}"),
    ("{user_email}", NULL, "schema_name", "default"),
    ("{user_email}", NULL, "paths.working_dir", "/Volumes/dbacademy/ops/dbacademy_{user_name_no_email}"),
    ("{user_email}", NULL, "cluster_name", "abc"),
    ("{user_email}", NULL, "warehouse_name", "shared_warehouse"),
    ("{user_email}", NULL, "pseudonym", "abc")
  ''')


insert_into_meta_source(user_email='infinitylearnings1201@gmail.com', 
                        user_name_no_email='infinitylearnings1201', 
                        catalog_name='dbacademy_peter_styliadis', 
                        schema_name='default')
            
insert_into_meta_source(user_email='mark.ott@databricks.com', 
                        user_name_no_email='mark_ott', 
                        catalog_name='dbacademy_mark_ott', 
                        schema_name='default')
```

```sql
SELECT * FROM dbacademy.ops.meta_source
```

```sql
CREATE OR REPLACE VIEW dbacademy.ops.meta AS
SELECT
  owner,
  object,
  key,
  value
FROM dbacademy.ops.meta_source
WHERE 
  CASE
    -- When a member of the admins group (true), you can see all rows
    WHEN owner = current_user() THEN TRUE
    -- Otherwise not part of the admins group (false), you can only see the IT rows
    ELSE FALSE
  END;
```
