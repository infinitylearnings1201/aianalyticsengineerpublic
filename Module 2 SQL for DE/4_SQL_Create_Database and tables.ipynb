# How to Create Databases & Tables

Up until now, we've been talking about databases in theory - how they are like Excel sheets, why SQL matters, and how data types give structure. But today we move to the first real act of building: creating a database and designing tables.

This is where things get real. A table is more than just a container for data; it is a disciplined structure that decides how information will be stored, retrieved, and trusted for years to come. 

For a data engineer, this step is as critical as laying the foundation of a building. Get it wrong, and the cracks will appear later in your pipelines, dashboards, and business decisions.

# Designing Tables - More Than Columns

When we design a table, we don't just say "here are the columns." We also define rules that enforce order:

- What kind of data belongs in each column (the data types we learned yesterday).
- Which column identifies each row uniquely.
- Which columns should never be duplicated.


These rules protect us from chaos. They make sure no two employees accidentally get the same ID, no two departments share the same name, and every employee record remains valid.

# Primary Key - The Identity of a Row

One of the first rules you set is the primary key. Think of it as the fingerprint of a record:

- It must be unique.
- It can never be empty.
- It guarantees that each row in the table can always be identified.

In our schema, the emp_id column in the employees table is the perfect primary key. No two employees should share the same ID, and every employee must have one. Without it, the database would struggle to tell Alice Johnson apart from another Alice Johnson.

Similarly, in the departments table, the dept_id is the natural primary key.

Here's how we create the departments table with a primary key:

## Uniqueness - Guardrails Against Bad Data

###  Why This Matters for Data Engineers


It's tempting to think creating tables is just boilerplate. But this step determines how trustworthy your data will be as it flows through the system. A pipeline is only as good as the quality of the data at its source.

- Without a primary key, you risk duplicate or missing records.
- Without uniqueness rules, you risk data that doesn't make sense (two employees sharing the same email).
- Without foreign keys, you risk orphan records (employees pointing to departments that don't exist).
