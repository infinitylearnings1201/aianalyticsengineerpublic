# SQL Joins — INNER, LEFT, RIGHT, FULL Explained Visually




In the real world, data rarely lives in one single table. Employees belong to departments. Orders belong to customers. Transactions link to products. To get meaningful insights, you often need to combine tables.

That’s what JOINs are for. They allow you to connect rows from two tables based on a shared column. In our case, the shared column is dept_id — it links employees to departments.

### 1. INNER JOIN
Visual: Only the overlap (like the intersection in a Venn diagram).
 It returns rows where the dept_id exists in both tables.

```sql
SELECT e.first_name, e.last_name, d.dept_name
FROM thedataengineering_00.company.employees e
INNER JOIN thedataengineering_00.company.department d
  ON e.dept_id = d.dept_id;
```

 Result: Every employee with a valid department.
 
 Excludes departments with no employees.

So if we had a department 5 with no employees, it would not appear here.

### 2. LEFT JOIN
Visual: All rows from the left table (employees), plus matches from the right table (departments).
 If no match, you still keep the employee, but department info will be NULL.

```sql
SELECT e.first_name, e.last_name, d.dept_name
FROM employees e
LEFT JOIN departments d
  ON e.dept_id = d.dept_id;
```

Result: Every employee is shown.
If an employee belongs to a department that doesn’t exist in departments, the department name will just show as NULL.

This is useful when you care about “all employees, even if their department is missing.”

### 3. RIGHT JOIN
Visual: The opposite of LEFT. All rows from the right table (departments), plus matching employees.

```sql
SELECT e.first_name, e.last_name, d.dept_name
FROM employees e
RIGHT JOIN departments d
  ON e.dept_id = d.dept_id;
```

Result: Every department is shown, even if it has no employees.
For example, if Marketing (dept_id = 4) had no employees, it would still appear, but employee columns would be NULL.

This is useful when you care about “all departments, even if no one is assigned yet.”

### 4. FULL OUTER JOIN
Visual: Everything from both sides.
 It combines LEFT and RIGHT — all employees, all departments, matched where possible, and NULLs where not.

```sql
SELECT e.first_name, e.last_name, d.dept_name
FROM employees e
FULL OUTER JOIN departments d
  ON e.dept_id = d.dept_id;
```

Result:

Employees with valid departments.

Employees without valid departments (still listed, dept = NULL).

Departments with no employees (still listed, employee = NULL).

This join is the most inclusive and ensures no data is left out.

### Why Joins Matter for Data Engineers
As a data engineer, JOINs are one of your most powerful tools. They allow you to:

Combine transactional data with reference data (e.g., orders + products).

Enrich raw logs with lookup tables.

Build dimensional models for analytics.

But they also require care — large JOINs can be expensive in distributed systems, and the wrong type of JOIN can change your results dramatically.

# Closing Thought

- INNER JOIN → Only matching rows.
- LEFT JOIN → All from left, matches from right.
- RIGHT JOIN → All from right, matches from left.
- FULL OUTER JOIN → Everything from both sides.

Think of JOINs as the way to connect dots across tables. Once you master them, you’ll unlock the ability to answer much more complex questions with SQL.
