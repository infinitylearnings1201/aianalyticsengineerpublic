### Views & Materialized Views




### Why Data Engineers Use Them

If you’ve ever built dashboards, you’ve probably noticed how the same queries get written again and again.
 You join the same tables, calculate the same KPIs, and apply the same filters — only to display the result in slightly different formats.

As datasets grow and queries become more complex, this repetition becomes costly — both for you and for the system.
 That’s where views and materialized views come in.
 They help you organize, reuse, and optimize SQL logic — turning repetitive queries into reusable, reliable building blocks.

### The Story Before Views
Imagine your manager asks for a report of each department showing:

- The number of employees
- The total salary spent
- The average salary per person

You write a query like this:

```sql
SELECT d.dept_name,
       COUNT(e.emp_id) AS employee_count,
       SUM(e.salary) AS total_salary,
       ROUND(AVG(e.salary), 2) AS avg_salary
FROM employees e
JOIN departments d
  ON e.dept_id = d.dept_id
GROUP BY d.dept_name;
```

The query works perfectly.
 But then, tomorrow, someone from finance asks for the same data in another dashboard.
 Next week, the HR team wants to add a filter by department location.
 Every time, you copy this query and tweak it.
 Soon you have multiple versions scattered across dashboards, notebooks, and pipelines — each doing the same work slightly differently.

This is the problem views solve.

###  What is a View?
A view is like saving a query result as a virtual table.
 You don’t store the data itself — just the SQL logic behind it.
 Whenever someone queries the view, Spark runs the underlying query dynamically and returns the latest results.

You can think of a view as a named query, or simply:

“A reusable SQL statement that behaves like a table.”

Here’s how to create one in Spark SQL:

```sql
CREATE OR REPLACE VIEW dept_salary_summary AS
SELECT d.dept_name,
       COUNT(e.emp_id) AS employee_count,
       SUM(e.salary) AS total_salary,
       ROUND(AVG(e.salary), 2) AS avg_salary
FROM employees e
JOIN departments d
  ON e.dept_id = d.dept_id
GROUP BY d.dept_name;
```

Now, instead of rewriting that entire query every time, you can just do:

```sql
SELECT * FROM dept_salary_summary;
```

The result is the same — but the logic is centralized, versioned, and easy to maintain.

### Why Views Are Useful
- Reusability — Once a view is defined, everyone can use it.
- Consistency — All teams rely on the same definition of metrics.
- Security — You can expose a view to users instead of the raw table (masking sensitive columns).
- Simplicity — Dashboards and reports become lightweight — no need for complex embedded SQL.

For a data engineer, views represent a semantic layer — they bridge raw data and analytics in a controlled, reliable way.

### What is a Materialized View?
A materialized view looks similar but behaves differently.
 While a regular view runs its query every time it’s accessed, a materialized view stores the results.

Think of it as a snapshot of the query output, saved as a table, which can be refreshed periodically.

In Spark SQL (and Databricks), you can create a materialized view like this:

```sql
CREATE MATERIALIZED VIEW dept_salary_summary_mv
AS
SELECT d.dept_name,
       COUNT(e.emp_id) AS employee_count,
       SUM(e.salary) AS total_salary,
       ROUND(AVG(e.salary), 2) AS avg_salary
FROM employees e
JOIN departments d
  ON e.dept_id = d.dept_id
GROUP BY d.dept_name;
```

Now, Spark doesn’t re-run the entire query each time.
 It reads directly from stored results, which makes repeated access faster and cheaper — especially for heavy analytical queries.

When new data comes in, you can refresh the materialized view to update it:

```sql
ALTER MATERIALIZED VIEW dept_salary_summary_mv REFRESH;
```

You can think of it like this:

A view is a recipe, and a materialized view is a prepared meal based on that recipe.

Why Data Engineers Love Materialized Views
In modern data pipelines, some queries — like daily KPIs or dashboards — run dozens of times a day.
 If each run scans millions of rows, that’s expensive and slow.

By materializing those results once, you:

- Save compute cost by not re-running complex joins or aggregations.
- Reduce latency for dashboards and APIs.
- Simplify data architecture by pre-staging data in a form ready for analytics.

Materialized views are especially powerful in lakehouse architectures like Databricks, where they integrate seamlessly with Delta tables and can be incrementally refreshed as new data arrives.

Combining the Two in Practice
A practical data engineer uses both:

- Views for logical organization (business definitions, reusable queries).
- Materialized views for performance optimization (cached results for analytics).

For example:

- Create a view employee_salary_summary for your team to access.
- Build a materialized view employee_salary_summary_mv behind it, refreshed daily.
- Your dashboard queries the view, but the heavy lifting is done by the materialized version.

This pattern keeps logic clean, cost low, and data fresh enough for daily analytics.

# Closing Thought
Views and materialized views are not just technical tools — they’re architectural patterns.
 They represent how we separate business logic from data storage, and how we build scalable, maintainable systems.

As a data engineer, your job isn’t just to process data; it’s to make it usable, performant, and trusted.
 Views bring order. Materialized views bring efficiency.
 Together, they turn SQL from a collection of queries into a well-structured data ecosystem.
