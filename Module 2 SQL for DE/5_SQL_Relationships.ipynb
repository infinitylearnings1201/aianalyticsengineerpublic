we built our first database and defined tables with primary keys and uniqueness rules. That gave each row a clear identity. But in the real world, data never lives in isolation. An employee belongs to a department. A department may manage multiple employees. These kinds of relationships are at the heart of database design.

If primary keys are about identity, then foreign keys are about connection. They ensure that one table can reliably point to another, and together they maintain data integrity.

# Why Relationships Matter

Imagine we have two tables: departments and employees.

- The departments table lists all valid departments in the company.
- The employees table lists individual employees.

```sql
%sql

select * from thedataengineering_00.company.employees
```

```sql
%sql

select * from thedataengineering_00.company.department
```

```sql
%sql

use catalog thedataengineering_00;
create schema relationships;
```

```sql
%sql
CREATE TABLE thedataengineering_00.relationships.departments (
    dept_id     INT PRIMARY KEY,
    dept_name   VARCHAR(100) NOT NULL,
    location    VARCHAR(100)
);
```

## Insert into Employees Table

```sql
%sql

CREATE TABLE If not exists thedataengineering_00.relationships.employees (
    emp_id      INT PRIMARY KEY,
    first_name  VARCHAR(50) NOT NULL,
    last_name   VARCHAR(50) NOT NULL,
    email       VARCHAR(100) NOT NULL,
    hire_date   DATE NOT NULL,
    salary      DECIMAL(10,2) NOT NULL,
    dept_id     INT NOT NULL,
    CONSTRAINT fk_employees_dept FOREIGN KEY (dept_id) REFERENCES departments(dept_id));
```
