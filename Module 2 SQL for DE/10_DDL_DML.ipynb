# DDL vs DML Explained Simply

Every data engineer eventually realizes this:

 Not all SQL statements are the same. Some build the structure of the database; others fill that structure with data.

 Understanding this difference is like understanding the blueprint versus the building - one designs, the other lives and breathes.

These two worlds of SQL are called DDL (Data Definition Language) and DML (Data Manipulation Language).

 They look similar at first glance, but their roles in a data engineer's life are very different.

The Two Sides of SQL

Think of your database as a house.
When you lay the foundation, build rooms, and decide how the wiring should run - that's DDL.

When you move in furniture, paint walls, or rearrange things - that's DML.

# DDL - Defining the Structure

DDL stands for Data Definition Language.

 These commands tell Spark (or any SQL engine) what the database should look like.

 You use DDL when you're creating, altering, or removing objects like tables, views, or databases.

Common DDL statements include:

- CREATE – to create new objects
- ALTER – to modify existing objects
- DROP – to delete them
- TRUNCATE – to remove all data but keep the structure

Let's look at some simple Spark SQL examples.

## Creating a Database

```sql
CREATE DATABASE IF NOT EXISTS company_db;
```

## Creating a Table

```sql
CREATE OR REPLACE TABLE employees (
  emp_id INT,
  first_name STRING,
  last_name STRING,
  dept_id INT,
  salary DOUBLE,
  hire_date DATE
);
```

## Altering a Table

```sql
ALTER TABLE employees ADD COLUMNS (email STRING);
```

## Dropping a table

```sql
DROP TABLE IF EXISTS employees;
```

# DML - Manipulating the Data

Once your tables are defined, it's time to work inside them.

 This is where DML (Data Manipulation Language) comes in.

 DML deals with inserting, updating, deleting, and reading records.

Common DML statements include:

- INSERT – to add data
- UPDATE – to change data
- DELETE – to remove data
- SELECT – to retrieve data

Let's see them in action.

## Inserting Data

```sql
INSERT INTO employees (emp_id, first_name, last_name, dept_id, salary, hire_date)
VALUES (101, 'Alice', 'Johnson', 1, 60000, '2020-01-15');
```

## Updating Data

```sql
UPDATE employees
SET salary = 65000
WHERE emp_id = 101;
```

## Deleting Data

```sql
DELETE FROM employees
WHERE dept_id = 4;
```

## Selecting Data

```sql
SELECT first_name, last_name, salary
FROM employees
WHERE salary > 70000;
```

For a data engineer, understanding DDL and DML isn't just academic - it's operational.

When building data pipelines or lakehouse architectures in Spark, you constantly switch between the two:

- You use DDL to design your schema - defining how tables, views, and databases connect.
- You use DML to load, transform, and analyze data flowing through those structures.

And here's the subtle but crucial distinction:
DDL affects the metadata - it changes your catalog, your definitions, your structure.

DML affects the data itself - it changes what's stored, but not the underlying design.

In a large team environment, these differences also translate into permissions:

Data engineers or admins often have DDL privileges (to create or alter objects).

Analysts and dashboard users usually get DML privileges (to query or modify data)

# Closing Thought

DDL and DML are the two languages every data engineer must be fluent in.
DDL builds the bones of your data system.
DML fills it with life.

Without DDL, there's nowhere to store data.
 Without DML, there's nothing worth storing.
Understanding when to use each - and how they complement each other - is what separates someone who "writes queries" from someone who designs data systems.
