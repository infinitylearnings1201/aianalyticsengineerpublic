# CASE WHEN & Conditional Logic in Spark SQL




So far, we’ve learned how to store, insert, update, delete, and filter data. But real-world data often demands more than just filtering — it requires logic.

For example, You may want to group employees into “High Salary” and “Low Salary” buckets. You may want to flag employees hired before 2020 as “Senior” and after 2020 as “Junior.”

In programming languages, we’d use if-else statements. In SQL, the equivalent is CASE WHEN.

### What is CASE WHEN?

The CASE WHEN expression allows you to apply conditional logic inside a query. You tell the database: if this condition is true, return this value; otherwise, return something else.

It works seamlessly inside a SELECT statement, making it incredibly useful for data transformations.

Example 1: Categorizing Salaries
Let’s divide employees into salary bands:

```python
SELECT emp_id, first_name, last_name, salary,
       CASE 
           WHEN salary >= 80000 THEN 'High Salary'
           WHEN salary BETWEEN 50000 AND 79999 THEN 'Medium Salary'
           ELSE 'Low Salary'
       END AS salary_band
FROM thedataengineering_00.company.employees;
```

- Salaries ≥ 80,000 → “High Salary”
- Between 50,000 and 79,999 → “Medium Salary”
- Below 50,000 → “Low Salary”

This instantly adds a new derived column, salary_band, to your results.

### Example 2: Senior vs Junior Employees

Suppose HR wants to know who joined before 2020 (senior) vs after (junior):

```python
%sql
SELECT first_name, last_name, hire_date,
       CASE 
           WHEN year(hire_date) < 2020 THEN 'Senior'
           ELSE 'Junior'
       END AS employee_level
FROM thedataengineering_00.company.employees;
```

### Example 3: Department Names Instead of IDs

Remember that in our employees table, we store only dept_id. Sometimes we want to display friendly department names without a join. CASE WHEN can help (though joins are usually better in production).

```python

SELECT first_name, last_name, dept_id,
       CASE 
           WHEN dept_id = 1 THEN 'HR'
           WHEN dept_id = 2 THEN 'Finance'
           WHEN dept_id = 3 THEN 'IT'
           WHEN dept_id = 4 THEN 'Marketing'
           ELSE 'Unknown'
       END AS department_name
FROM thedataengineering_00.company.employees;
```

### Example 4: Flagging Special Conditions

Maybe we want to quickly identify employees with missing email addresses:

```python
%sql
SELECT 
    first_name, 
    last_name,
    CASE 
        WHEN last_name IS NULL THEN 'Missing Last Name'
        ELSE 'Valid Last Name'
    END AS email_status
FROM thedataengineering_00.company.employees;
```

### Why CASE WHEN Matters for Data Engineers

For a data engineer, CASE WHEN is not just about pretty labels — it’s about transforming raw data into business-friendly insights.

It allows you to derive new columns without altering tables.

It adds categorization and flags directly in SQL, reducing the need for downstream logic.

It’s often used in data quality checks (e.g., missing values, invalid ranges).

It’s critical for reporting pipelines, where business teams expect clean categories like “Gold Customer,” “Silver Customer,” “Bronze Customer.”

In Spark SQL, CASE WHEN scales effortlessly — whether you’re working with 10 rows or 10 billion.
