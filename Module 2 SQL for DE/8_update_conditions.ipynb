# UPDATE with Conditions

Databases are living systems. Employees join, leave, or get promoted. Departments change names or locations. Salaries get revised. If all we ever did was insert data, the database would quickly become outdated. That's why SQL gives us the UPDATE statement - the way to modify existing records without creating duplicates.

But with this power comes responsibility. An update can be precise and surgical, or it can be destructive if done carelessly. The key is learning how to update safely, with conditions.

# The Role of UPDATE

Unlike INSERT, which adds new rows, UPDATE changes values in rows that already exist. But here's the important detail: if you run an UPDATE without conditions, it will change every row in the table.

That's why conditions are so critical. They tell the database which specific rows should be updated.

```sql
UPDATE thedataengineering_00.company.employees
SET salary = 80000
WHERE emp_id = 102;
```

# Updating Based on Conditions


You can also update multiple rows at once by setting conditions.

```sql
select * from thedataengineering_00.company.employees where dept_id = 3
```

```sql
UPDATE thedataengineering_00.company.employees
SET salary = salary * 1.05
WHERE dept_id = 3;
```

```sql
select * from thedataengineering_00.company.employees where dept_id = 3
```

# Updating Multiple Columns

Sometimes you need to modify more than one piece of information at once.

```sql
select * from thedataengineering_00.company.employees where emp_id = 105
```

```sql
UPDATE thedataengineering_00.company.employees
SET last_name = 'Williams',
    salary = 50000
WHERE emp_id = 105;
```

```sql
select * from thedataengineering_00.company.employees where emp_id = 105
```

# Avoiding Mistakes

The danger with UPDATE is forgetting the WHERE clause or writing it incorrectly. Always remember:

- Without WHERE, every row is updated.
- With the wrong condition, you might update the wrong set of rows.


A safe practice is to first run a SELECT with the same condition to see which rows will be affected.

For data engineers, UPDATE is more than just fixing typos or changing salaries. It's part of keeping data fresh, accurate, and trustworthy. You might:

- Correct errors in ingested data.
- Adjust values when upstream systems change.
- Apply transformation logic directly in staging tables.


But because updates can alter large amounts of data in seconds, they must be done with discipline. Always double-check the conditions, and in production systems, updates are often logged or run within transactions to ensure safety.

# Closing Thought

INSERT introduced life into your tables. SELECT gave you the power to ask questions. Now, UPDATE gives you the ability to evolve your data over time. It's a reminder that databases are not static - they change as the real world changes.

Handled with care, UPDATE is your ally in keeping data aligned with reality. Handled recklessly, it can wipe out accuracy in seconds. As a data engineer, that's why you must always treat UPDATE with respect.
