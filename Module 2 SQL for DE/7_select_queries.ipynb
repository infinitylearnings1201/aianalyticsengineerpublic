# Reading Data in Different Ways

By now, our departments and employees tables are no longer empty blueprints. They have records - names, salaries, departments, hire dates. But a database filled with data is useless until we can read it back in meaningful ways. That's what the SELECT statement is for.

If INSERT is about writing a story into the database, SELECT is about reading it back. This is the SQL command you will use most often, not just as a beginner but throughout your career.

# The Simplest Form

The most basic way to use SELECT is to return everything from a table.

```sql
SELECT * FROM thedataengineering_00.company.employees;
```

Selecting with column names

```sql
SELECT first_name, last_name, salary FROM thedataengineering_00.company.employees;
```

# Filtering with WHERE

```sql
SELECT first_name, last_name 
FROM thedataengineering_00.company.employees
WHERE dept_id = 2;
```

```sql
SELECT first_name, last_name, salary
FROM thedataengineering_00.company.employees
WHERE salary > 70000;
```

# Filtering with Ranges

```sql
SELECT first_name, last_name, hire_date
FROM thedataengineering_00.company.employees
WHERE hire_date BETWEEN '2019-01-01' AND '2021-12-31';
```

# Filtering with Lists

```sql
SELECT first_name, last_name, dept_id
FROM thedataengineering_00.company.employees
WHERE dept_id IN (1, 3);
```

# Combining Conditions

```sql
SELECT first_name, last_name, salary
FROM thedataengineering_00.company.employees
WHERE dept_id = 3 AND salary > 60000;
```

```sql
SELECT first_name, last_name, dept_id, salary
FROM thedataengineering_00.company.employees
WHERE dept_id = 1 OR salary > 90000;
```

# Sorting Results

Databases don't guarantee any natural order of rows. If you want results sorted, you use ORDER BY.

```sql
SELECT first_name, last_name, salary
FROM thedataengineering_00.company.employees
ORDER BY salary DESC;
```

```sql
SELECT first_name, last_name, hire_date
FROM thedataengineering_00.company.employees
ORDER BY hire_date ASC;
```

# Aliases - Friendlier Output

Sometimes column names are too technical. You can rename them in the output using AS.

```sql
SELECT 
first_name AS name,
last_name  AS surname,
salary     AS compensation
FROM thedataengineering_00.company.employees;
```

# Closing Thought

By mastering SELECT, you've unlocked the ability to interrogate data. Instead of scrolling through endless rows like in Excel, you ask precise questions - "show me employees hired between these dates," "list everyone earning above this amount," "give me the top earners in Finance."

This is the moment SQL starts to feel magical: with just a few lines, you can sift through thousands or millions of rows and instantly see only what matters.
