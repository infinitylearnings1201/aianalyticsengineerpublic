#SQL Functions 

## Numeric & Date Functions (with Spark SQL Examples)

Numbers and dates are central to most business data. Salaries, employee IDs, and hire dates are not just stored, they are calculated, compared, and analyzed every day. Spark SQL, just like other SQL engines, provides a wide range of numeric and date functions to work with this data efficiently.

Let's walk through them with simple examples on our familiar employees table.

##  1. ROUND - Rounding Numbers

```sql
SELECT emp_id, salary, ROUND(salary, -3) AS rounded_salary
FROM thedataengineering_00.company.employees;
```

## 2. CEIL and FLOOR - Always Up or Down

```sql
SELECT emp_id, salary,
       CEIL(salary/1000.0) * 1000 AS ceil_salary,
       FLOOR(salary/1000.0) * 1000 AS floor_salary
FROM thedataengineering_00.company.employees;
```

## 3. ABS - Absolute Value

```sql
SELECT ABS(-5000) AS positive_value;
```

## 4. Arithmetic in Queries

```sql
SELECT emp_id, salary,
       salary * 1.10 AS new_salary   -- 10% increment
FROM thedataengineering_00.company.employees;
```

# Date Functions in Spark SQL

## Current Date

```sql
SELECT current_date() AS today
```

## Extracting Year, Month, Day

```sql
SELECT emp_id, hire_date,
       year(hire_date)  AS hire_year,
       month(hire_date) AS hire_month,
       day(hire_date)   AS hire_day
FROM thedataengineering_00.company.employees;
```

# Date Difference

To calculate how many days an employee has been in the company:

```sql
SELECT emp_id, first_name, hire_date,
       datediff(current_date(), hire_date) AS days_in_company
FROM thedataengineering_00.company.employees;
```

## Add Days or Months

```sql
SELECT emp_id, hire_date,
       date_add(hire_date, 90)    AS after_90_days,
       add_months(hire_date, 12) AS after_1_year
FROM thedataengineering_00.company.employees;
```

## Truncating Dates (to Month/Year)

```sql
SELECT emp_id, hire_date,
       trunc(hire_date, 'MM') AS month_start,
       trunc(hire_date, 'YYYY') AS year_start
FROM thedataengineering_00.company.employees;
```

In real-world pipelines, data rarely comes perfectly formatted. You may need to:
Put salaries into ranges (numeric functions).

- Standardize increments (arithmetic).
- Calculate tenure or filter by hire dates (date functions).
- Aggregate over months or years (truncation).

With Spark SQL, these operations are not just quick but also scalable - you can run the same functions on thousands or billions of rows.
