# SQL Functions - String Operations (Concatenation, Trimming, etc.)

## Concatenation - Joining Strings Together

```python
%sql
SELECT first_name || ' ' || last_name AS full_name
FROM thedataengineering_00.company.employees;
```

```python
%sql
SELECT TRIM(first_name) AS clean_first_name
FROM thedataengineering_00.company.employees;
```

# UPPER and LOWER - Consistency in Case

```python
%sql
SELECT UPPER(dept_name) AS department
FROM thedataengineering_00.company.department;
```

```python
%sql
SELECT LOWER(dept_name) AS normalized_email
FROM thedataengineering_00.company.department;
```

## LENGTH - Measuring Text

```python
%sql
SELECT first_name, LENGTH(first_name) AS name_length
FROM thedataengineering_00.company.employees;
```

## SUBSTRING - Extracting Parts of Text

```python
%sql
SELECT SUBSTRING(first_name, 1, 3) AS first_five_chars
FROM thedataengineering_00.company.employees;
```

## Combining Functions

```python
%sql

SELECT UPPER(CONCAT(TRIM(first_name), ' ', TRIM(last_name))) AS clean_full_name
FROM thedataengineering_00.company.employees;
```
