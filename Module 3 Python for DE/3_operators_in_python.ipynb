# Operators & Expressions in Python

### Why This Topic Matters
Every data transformation you perform - from calculating sales growth to comparing timestamps or filtering rows - ultimately uses operators and expressions.

Whether you're writing:

- a filter condition (if revenue > 10000:),
- a derived column (profit = revenue - cost),
- or a data validation rule (if price == 0 or quantity == 0:),

you are using operators.
Understanding how they behave is the difference between a pipeline that works and one that silently produces wrong results.

# What Are Operators?

Operators are special symbols or keywords that perform operations on values (also called operands).

## Operators + Operands = Expressions

```python
#Lets take an example of a subtraction operation

revenue = 12000
cost = 7000

profit = revenue - cost

print(profit)
```

```python
price = 10
qty = 100

total_sales = price * qty

print(total_sales)
```

### Comparison Operator

==, !=, >, <, >=, <=>

```python
revenue = 12000
target = 100000

print("Target achieved?", revenue >= target)
```

### Logical Operators

Used to combine multiple conditions - often inside if statements.

and, or, not

```python
revenue = 12000
cost = 7000

if(revenue > 10000 and cost < 8000):
  print("Profitable")
else:
    print("Not profitable")
```

# Assignment Operators

Used to store results in variables.

- = Assignment Operator
- += Add and assign
- -= Subtract and assign
- *= multiply and assign
- /= Divide and assign
%= modulus and assign

```python
records = 100

records += 100 (record = record + 100)

print(records)
```

# Combining Expressions

You can chain multiple operations together:

```python
revenue = 100000
cost = 75000
profit_margin = (revenue - cost) / revenue * 100

print("Profit Margin:", profit_margin, "%")
```

This kind of calculation appears everywhere in data pipelines - think of it as the Python equivalent of a SQL computed column.

```python
test = 100 + 100*2
print(test)
```

```python
test = (100 + 100)*2
print(test)
```

# Operator Precedence

Python evaluates expressions from left to right but respects operator priority.
 Parentheses () always come first.

Order of precedence:

- ()  Brackets
- **  Orders
- *, /, //, %   Division, Multiplication
- +, -       Addition, Substraction
- Comparison (>, <, etc.)
- Logical (and, or, not)

(BODMAS - Brackets, Orders, Division, Multiplication, Addition, Substraction)

```python
result = 10 + 5 * 2
print(result)     # 20 because * runs before +
```

```python
result = (10 + 5) * 2
print(result)     # 20 because * runs before +
```

```python
result = (10 + 5) ** 2
```

##  Mini Challenge - Eligibility for Reward

```python
salary = 80000
bonus = 5000
tax_rate = 0.18

# Calculate net salary after tax
net_salary = (salary + bonus) - (salary * tax_rate)
print("Net Salary:", net_salary)

# Check if eligible for performance reward
if net_salary > 70000 and bonus > 0:
    print("Eligible for reward!")
else:
    print("Not eligible.")
```

# Summary

Operators are building blocks for all expressions and conditions.

- Arithmetic → calculations
- Comparison → conditions
- Logical → combine conditions
- Assignment → store results

Together, they help data engineers transform, compare, and validate data efficiently - whether inside a pandas transformation or an ETL function.
