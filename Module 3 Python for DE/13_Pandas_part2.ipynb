# Pandas Part 2 - Filtering, Sorting, and Selecting Columns

In data engineering, you often:
  
- Filter data based on conditions (e.g., WHERE salary > 10000)
- Sort results (ORDER BY salary DESC)
- Select only certain columns for downstream pipelines

In pandas, these operations are simple, readable, and chainable, allowing you to express SQL-like logic entirely in Python.

### Let's Load Our Sample Tables

```python
import pandas as pd

# Employees table
employees_df = pd.DataFrame({
    "emp_id": [1, 2, 3, 4, 5],
    "emp_name": ["Akash", "Priya", "Rohan", "Neha", "Raj"],
    "department_id": [10, 20, 30, 10, 20],
    "salary": [12000, 8000, 15000, 11000, 9000]
})

# Department table
department_df = pd.DataFrame({
    "department_id": [10, 20, 30],
    "department_name": ["Sales", "Marketing", "IT"]
})
```

### Selecting Columns - Like SQL SELECT

```python
employees_df["emp_name"]
```

```python
employees_df[["emp_name", "salary"]]
```

It selects all rows (:) and only the 2nd and 4th columns (index positions [1, 3]) from the DataFrame. In this case, it returns the emp_name and salary columns.

In pandas, .loc[] is a label-based selector used to access rows and columns by their names (labels) rather than index numbers.

column -> 0, 1,2,3

```python
employees_df.iloc[2:5, [0,2]]
```

### Filtering Rows - Like SQL WHERE

```python
print(employees_df["salary"])
```

```python
high_salary = employees_df[employees_df["salary"] > 10000]
print(high_salary)
```

### Multiple Conditions with & and |

```python
filtered = employees_df[
    (employees_df["salary"] > 9000) & 
    (employees_df["department_id"] == 10)
]

print(filtered)
```

### Using .isin() for "IN" Condition

```python
filtered = employees_df[employees_df["department_id"].isin([10, 30])]

print(filtered)
```

```python
filtered = employees_df[employees_df["emp_name"].str.startswith("R")]

print(filtered)
```

### Sorting Rows - Like SQL ORDER BY

```python
sorted_df = employees_df.sort_values(by="salary", ascending=False)
print(sorted_df)
```

```python
sorted_multi = employees_df.sort_values(by=["department_id", "salary"], ascending=[True, False])
print(sorted_multi)
```

### Selecting Columns After Filter

```python
employees_df.loc[employees_df["salary"] > 10000, ["emp_name", "salary"]]
```

### Merging Back with Departments

```python
merged_df = employees_df.merge(department_df, on="department_id", how="left")
filtered_df = merged_df[merged_df["department_name"] == "Sales"]
print(filtered_df)
```

### Chaining Operations (ETL Style)

You can combine filters, sorting, and selection in one flow:

```python
result = (
    employees_df
    .query("salary > 9000")
    .sort_values("salary", ascending=False)
    [["emp_name", "salary"]]
)
print(result)
```
