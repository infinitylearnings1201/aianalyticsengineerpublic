# Loops in Python - for & while

In data engineering, we process thousands - or millions - of records every day. 

When you:

- Read each row from a CSV,
- Validate a list of file names,
- Or calculate totals from multiple data entries,

you're using loops behind the scenes.
Loops are how you iterate - move through a dataset, one element at a time - and apply logic.

# What Is a Loop?

A loop allows you to repeat a block of code as long as a condition is true, or for every item in a collection.

Python has two main types of loops:

- for loop - iterate over items
- while loop - repeat until a condition fails

## 1. The for Loop

The for loop is used when you know what you're looping over - like a list, dictionary, or range of numbers.

        for variable in sequence:
                 # do something

```python
regions = ["US", "EU", "APAC"]

for r in regions:
    print("Processing region:", r)
```

```python
for i in range(1,10):
  print(i)
```

## Example: Looping Through a Dataset

```python
sales = [12000, 8000, 15000, 6000]

for amount in sales:
    if amount > 10000:
        print("High sale:", amount)
    else:
        print("Normal sale:", amount)
```

## Looping Through Dictionaries

```python
record = {"region": "US", "revenue": 12000, "cost": 8000}

for key, value in record.items():
    print(key, "→", value)
```

# 2. The while Loop

Used when you don't know in advance how many times to loop - you keep looping while a condition is true.

    while condition
        #do something


```python
count = 1
while count <= 5:
    print("Processing batch:", count)
    count += 1
```

# Using break and continue

These control how loops behave: break → Stop the loop completely

continue → Skip current iteration, go to next

```python
for num in range(1, 6):
    if num == 3:
        continue
    print(num)
```

## Data Validation Loop

```python
files = ["sales_jan.csv", "sales_feb.csv", "temp.txt", "sales_mar.csv"]

for f in files:
    if not f.endswith(".csv"):
        print(f, "→ Skipped (Invalid format)")
        continue
    print(f, "→ Ready for ingestion") -> Ingest_File()
```

## Nested Loops

```python
datasets = {
    "US": [12000, 15000],
    "EU": [8000, 7000]
}

for region, sales in datasets.items():
    for s in sales:
        print(region, "→", s)
```

## Loop + Condition - A Classic Data Engineering Pattern

```python
records = [
    {"id": 1, "amount": 12000},
    {"id": 2, "amount": 8000},
    {"id": 3, "amount": 5000},
]

for r in records:
    if r["amount"] > 10000:
        print(" Valid:", r)
    else:
        print(" Skipped:", r)
```
