# What Is a Function? 

A function is a named block of code that performs a specific task and can be reused anywhere in your program.

        def function_name(parameters):
            # logic
            return result

```python
def greet_user():
    print("Welcome to Python for Data Engineers! Let's get started!!!!!")
```

```python
greet_user()
```

```python
for i in range(1,6):
    greet_user()
```

# 1. Functions with Parameters

Parameters make your function flexible.

```python
def calculate_profit(revenue, cost):
    profit = revenue - cost
    return profit
```

```python
p = calculate_profit(15000, 9000)
print("Profit:", p)
```

# 2. Default Parameters

You can define default values for parameters.

```python
def calculate_tax(amount, rate=0.18):
    return amount * rate
```

```python
print(calculate_tax(10000))       # Uses default rate
print(calculate_tax(10000, 0.25)) # Override rate
```

# 3. Returning Multiple Values

You can return multiple results as a tuple.

```python
def compute_metrics(revenue, cost):
    profit = revenue - cost
    margin = (profit / revenue) * 100
    return profit, margin

p, m = compute_metrics(15000, 9000)
print("Profit:", p)
print("Margin:", m, "%")
```

# 4. Using Functions in Loops

Combine what you learned earlier - loops + functions = scalable logic.

```python
def evaluate_sale(amount):
    if amount > 10000:
        return "High"
    else:
        return "Normal"

sales = [8000, 12000, 9500, 15000]

for s in sales:
    print(s, "→", evaluate_sale(s))
```

# 5. Keyword & Positional Arguments

You can pass parameters by position or by name.

```python
def describe_customer(name, country):
    print(f"Customer: {name}, Country: {country}")

# Positional
describe_customer("John", "India")

# Keyword
describe_customer(country="US", name="Jane")
```

# 6. *args and **kwargs

Useful when you don't know how many inputs you'll get.

### *args → many positional arguments

```python
def total_amount(*values):
    return sum(values)

print(total_amount(100, 200, 300, 400))
```

### **kwargs → many keyword arguments

```python
def print_details(**info):
    for key, value in info.items():
        print(key, ":", value)

print_details(name="John", role="Data Engineer", country="India", city="Bangalore", salary=50000)
```

### Data Validation Function

```python
def validate_record(record):
    if not record.get("id") or record.get("amount", 0) <= 0:
        return False
    return True

records = [
    {"id": 1, "amount": 5000},
    {"id": 2, "amount": 0},
    {"id": None, "amount": 8000}
]

for r in records:
    if validate_record(r):
        print("Valid Record:", r)
    else:
        print("Invalid Record:", r)
```

# Functions Returning Other Functions (Advanced Peek)

You can even define a function that creates another function - a foundation for modular frameworks and decorators.

```python
def multiplier(factor):
    def multiply(value):
        return value * factor
    return multiply

multiply_me = multiplier(3)
print(multiply_me(20))   # 20
```

# Build Your Own ETL Step

```python
def extract(): 
    return [12000, 8000, 15000, 60000]

def transform(sales):
    return [s for s in sales if s > 10000]

def load(filtered):
    print("Loaded records:", len(filtered))

data = extract()
filtered = transform(data)
load(filtered)
```
