# Pandas Part 3 - GroupBy, Aggregations, and Joins/Merges

In real-world data pipelines, data engineers frequently:

- Summarize transactions per customer or department
- Compute metrics (sums, averages, counts)
- Combine multiple datasets into one unified view

In pandas, this is done with groupby(), aggregate(), and merge() - the same concepts behind SQL's analytical processing.

### Lets Build Sample Tables

```python
import pandas as pd

employees_df = pd.DataFrame({
    "emp_id": [1, 2, 3, 4, 5, 6],
    "emp_name": ["Akash", "Priya", "Rohan", "Neha", "Raj", "Amit"],
    "department_id": [10, 20, 30, 10, 20, 30],
    "salary": [12000, 8000, 15000, 11000, 9000, 17000]
})

department_df = pd.DataFrame({
    "department_id": [10, 20, 30],
    "department_name": ["Sales", "Marketing", "IT"]
})
```

### Intro to groupby() - The SQL GROUP BY Equivalent

```python
employees_df.groupby("department_id").agg({"salary": "mean"})
```

### Multiple Aggregations

```python
dept_summary = (
    employees_df
    .groupby("department_id")
    .agg(
        emp_count=("emp_id", "count"),
        avg_salary=("salary", "mean"),
        total_salary=("salary", "sum")
    )
    .reset_index()
)
print(dept_summary)
```

### Grouping by Multiple Columns

```python
grouped = employees_df.groupby(["department_id", "emp_name"]).size().reset_index(name="count")
print(grouped)
```

### Aggregating Multiple Columns at Once

```python
salary_summary = (
    employees_df.groupby("department_id")[["salary"]]
    .agg(["min", "max", "mean"])
)
print(salary_summary)
```

### Joining Grouped Results with Department Names

```python
summary_df = (
    employees_df
    .groupby("department_id")
    .agg(emp_count=("emp_id", "count"), avg_salary=("salary", "mean"))
    .reset_index()
    .merge(department_df, on="department_id", how="left")
)
print(summary_df)
```

### Using transform() for Per-Row Calculations

```python
employees_df["dept_avg_salary"] = employees_df.groupby("department_id")["salary"].transform("mean")
print(employees_df)
```

### Combining GroupBy + Sorting
You can quickly identify top earners per department.

```python
top_earners = (
    employees_df.sort_values(["department_id", "salary"], ascending=[True, False])
    .groupby("department_id")
    .head(1)
)
print(top_earners)
```

### Joining Multiple DataFrames
If you have more datasets (e.g., employee + department + region), pandas supports multiple joins.

```python
# Example: Left join employees → department
merged = employees_df.merge(department_df, on="department_id", how="left")
print(merged.head())
```
