### 1.2 Conditional Statements

```python
# Using if-else conditions
age = 10

if age > 25:
    print("Eligible for senior role")
else:
    print("Entry-level role")
```

```python
salary = 95000

if salary > 100000:
    risk = "High"
elif 50000 <= salary <= 100000:
    risk = "Medium"
else:
    risk = "Low"

print(f"Salary: ₹{salary} → Risk Category: {risk}")
```
