# File Handling Part 1 - Reading & Writing Text and CSV Files

Every data pipeline starts and ends with files - logs, CSVs, JSONs, or flat files. You'll read from a raw source, clean or transform it, and write it back.

As a data engineer, you'll often:
- Read raw CSVs from ingestion folders
- Write cleaned datasets to processed zones
- Log pipeline events into text files

Today we'll master text & CSV file operations - safely and efficiently.

###  1. Opening and Closing Files

###Why we use Databricks Volumes for practicals?

```python
file = open("/Volumes/thedataengineering_00/data/sampledata/sampleimages/sample.txt", "r")  # 'r' = read mode
content = file.read()
print(content)
file.close()
```

### 2. Using with open() - The Safe Way
This automatically closes the file, even if an error occurs.

```python
with open("/Volumes/thedataengineering_00/data/sampledata/sampleimages/sample.txt", "r") as f:
    data = f.read()
    print(data)
```

### 3. Reading Files Line by Line
Sometimes you want to process each record individually.

```python
with open("/Volumes/thedataengineering_00/data/sampledata/sampleimages/sample.txt", "r") as f:
    for line in f:
        print(line.strip() + "   awesome")
```

### Opportunity to apply conditions

```python
with open("/Volumes/thedataengineering_00/data/sampledata/sampleimages/sample.txt", "r") as f:
    for line in f:
        if line.startswith("Hello"):
            print(line.strip() + "   awesome")
```

### 4. Writing to a File
Use 'w' (write) or 'a' (append) modes.

```python
with open("/Volumes/thedataengineering_00/data/sampledata/sampleimages/log.txt", "w") as f:
    f.write("Data ingestion started.\n")
    f.write("File validated successfully.\n")
```

```python
with open("/Volumes/thedataengineering_00/data/sampledata/sampleimages/log.txt", "r") as f:
    for line in f:
        print(line.strip())
```

'w' overwrites existing content, 'a' appends to existing files.

### 5. Reading and Writing CSV Files
CSV (Comma-Separated Values) is the bread-and-butter of data exchange.

```python
import csv

header = ["id", "name", "amount"]
rows = [
    [1, "Akash", 12000],
    [2, "Priya", 8000],
    [3, "Rohan", 15000],
    [4, "Neha", 18000],
    [5, "Raj", 9500]
]
with open("/Volumes/thedataengineering_00/data/sampledata/sampleimages/python_sales.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(rows)
```

```python
import csv

with open("/Volumes/thedataengineering_00/data/sampledata/sampleimages/python_sales.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

### 6. Using DictReader and DictWriter
Prefer dictionaries for column-name access.

```python
import csv

with open("/Volumes/thedataengineering_00/data/sampledata/sampleimages/python_sales.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(row)
        print(row["name"], "→", row["amount"])
```

### Writing as Dictionaries

```python
data = [
    {"id": 4, "name": "Neha", "amount": 18000},
    {"id": 5, "name": "Raj", "amount": 9500}
]

with open("/Volumes/thedataengineering_00/data/sampledata/sampleimages/python_sales.csv", "a", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["id", "name", "amount"])
    for row in data:
        writer.writerow(row)
```

```python
import csv

with open("/Volumes/thedataengineering_00/data/sampledata/sampleimages/python_sales.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(row["name"], "→", row["amount"])
```

### 7. Handling File Errors Gracefully
Always use try-except around file operations.

```python
try:
    with open("missing.csv", "r") as f:
        data = f.read()
except FileNotFoundError:
    print(" File not found!")
```

### 8. Mini Data Engineering Scenario
Let's combine everything to read a raw CSV and write a clean version.

```python
import csv

with open("/Volumes/thedataengineering_00/data/sampledata/sampleimages/python_sales.csv", "r") as infile, open("/Volumes/thedataengineering_00/data/sampledata/sampleimages/clean_sales.csv", "w", newline="") as outfile:
    reader = csv.DictReader(infile)
    writer = csv.DictWriter(outfile, fieldnames=["id", "name", "amount"])
    writer.writeheader()

    for row in reader:
        row["name"] = row["name"].strip().title()
        writer.writerow(row)

print(" Clean file written successfully!")
```
