# Foundation Python Day 2 - Variables & Data Types

When we begin learning Python, our first goal isn't to memorize syntax - it's to learn how to think in terms of data.

 Every program you'll ever write - whether it cleans millions of rows in a data lake or loads an API into a warehouse - begins with one simple idea:

### "Store something, name it meaningfully, and use it purposefully."

That's what variables and data types are all about.

## Variables - Naming the Data You Work With

A variable is like a labeled container on your data engineer's workbench.

You decide what to store in it - a number, a name, a list, a table - and Python remembers it for you.

```python
# A few examples of variables
company_name = "Acme"
num_employees = 500
revenue_in_millions = 20.5
is_partnered_with_databricks = True
```

```python
# Print what is stored

print(company_name)
print(num_employees)
print(revenue_in_millions)
print(is_partnered_with_databricks)
```

Each of these is a tiny unit of memory holding some data. Notice how natural the naming feels - company_name, num_employees, revenue_in_millions.

 Good naming is the first sign of a thoughtful engineer.

# Data Types - The Nature of the Value Inside

Every variable has a type, just like every column in a database table.
If you've ever seen a schema like this:

You already understand data types - they describe what kind of data lives in each variable.

Let's explore the most common ones you'll use daily as a data engineer.

# Integers (int)

Whole numbers - no decimals. Used for counts, IDs, and indexes.

```python
order_count = 120
year = 2025

print(type(order_count))
print(type(year))
```

Whenever you're counting records, customers, or rows - you're dealing with integers.

# Floating-Point Numbers (float)

Numbers with decimals - used for precision values like revenue, percentages, or averages.

```python
average_order_value = 245.75
discount_rate = 0.15
```

If you mix integers and floats in an expression, Python automatically promotes the result to a float:

```python
total_revenue = order_count * average_order_value

print(total_revenue)
print(type(total_revenue))
```

This is why you can safely perform mathematical operations without worrying too much about precision - Python handles it gracefully for you.

# Strings (str)

Strings hold text - everything inside quotes.
 In data terms, this could be customer names, categories, or IDs.

```python
customer_name = "Akash W"
region = 'India'
```

Strings can be combined (concatenated) easily:

```python
greeting = "Hello, " + customer_name
print(greeting)
# Output: Hello, Akash W
```

They're also the most common data type in ETL work - think of all the product names, city codes, or log messages you handle every day.

# Booleans (bool)

Booleans represent truth values - True or False.
 They help you control logic, filter data, or validate rules.

```python
is_active_customer = True
has_unpaid_invoices = False
```

You'll use booleans extensively while building conditions and data quality checks.

# Lists (list)

A list is an ordered collection of items.
 Imagine a to-do list or a queue of datasets to process - that's what a list feels like.

```python
countries = ["India", "USA", "Brazil", "Germany"]
print(countries[0])   # Access first element

print(countries)
```

Lists are mutable, meaning you can add or remove elements.

```python
countries.append("Australia")
print(countries)
```

# Tuples (tuple)

Tuples are like lists, but immutable - once created, they cannot change.
 That's perfect for storing fixed configurations, constants, or coordinates.

```python
db_config = ("server01", "1433", "read_only_user")
print(db_config[0])  # Access first element
```

Tuples keep your constants safe from accidental modification during code execution.

# Dictionaries (dict)

Dictionaries store data as key-value pairs - like a small lookup table or configuration map.

```python
customer = {
    "id": 101,
    "name": "Ravi Sharma",
    "city": "Pune",
    "loyalty_points": 1200
}

print(customer)
```

They're incredibly useful when working with structured or JSON-like data - which you'll encounter everywhere in APIs and logs.

They're incredibly useful when working with structured or JSON-like data - which you'll encounter everywhere in APIs and logs.

To access data, you use the key:

```python
print(customer["city"])  # Output: Pune
```

### In a data engineering pipeline, dictionaries are often used for:
    
- Holding file mappings
- Storing metadata
- Passing configuration to functions

# Sets (set)

Sets are unordered collections of unique elements.

 They help you find distinct values or eliminate duplicates - an essential data-cleaning step.

```python
customer_ids = [101, 102, 103, 101, 102]
unique_customers = set(customer_ids)
print(unique_customers)
# Output: {101, 102, 103}
```

You'll appreciate sets when deduplicating IDs or comparing lists of values between systems.

### Why Understanding These Types Matters

In your future as a data engineer, you'll work with structured and unstructured data - often messy, inconsistent, and huge in scale.
 The first step to taming that chaos is knowing what kind of data you're dealing with and what operations it allows.

- Numbers can be aggregated.
- Strings can be cleaned or split.
- Lists can be iterated.
- Dictionaries can be mapped.
- Sets can be deduplicated.

The stronger your understanding of data types, the cleaner your transformations and the fewer your debugging nightmares.
