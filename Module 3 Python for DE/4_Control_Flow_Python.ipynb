# Control Flow in Python - if, elif, else & Conditions

Why Control Flow Matters in Data Engineering
Every pipeline needs decisions.

Think of some examples:

- If a file arrives in the bucket → trigger ingestion
- If revenue > cost → mark profit as "Positive"
- If column is null → replace it with a default value
- If data quality score < 90 → send alert

All of these come down to control flow - Python's way of letting your code "decide" what to do next.

# What Is Control Flow?

![](/Volumes/thedataengineering_00/data/sampledata/sampleimages/flow_diagram.png)

Control flow simply means controlling the order in which your code executes, based on certain conditions.

In Python, we use three main keywords for this:

-  if
-  elif (else if)
-  else

```python
revenue = 12000
cost = 8000

if revenue > cost:
    print("Profit achieved!")
```

### Adding else - Two-Way Decision

```python
revenue = 7000
cost = 9000

if revenue > cost:
    print("Profit achieved!")
else:
    print("Loss recorded!")
```

### Adding elif - Multiple Conditions

```python
revenue = 100000
cost = 1000000

if revenue > cost:
    print("Profit achieved!")
elif revenue == cost:
    print("Break-even point!")
else:
    print("Loss recorded!")
```

###  Combining Conditions with Logical Operators

```python
revenue = 15000
cost = 9000
region = "APAC"

if revenue > cost and region == "APAC":
    print("Strong APAC performance!")
```

### Nested if Statements

```python
country = "India"
sales = 12000

if country == "India":
    if sales > 10000:
        print("Top performer in India!")
```

But be careful: deep nesting makes code harder to read. Prefer and/or instead when possible.

### Realistic Example: Data Quality Check

In real data engineering, you'll often perform data validations before loading data.

```python
record_count = 10000
null_percentage = 3.5

if record_count == 0:
    print("No data received!")
elif null_percentage > 10:
    print("Data quality issue - too many nulls!")
else:
    print("Data looks good to load.")
```

### Example: Categorizing Revenue

```python
revenue = 45000

if revenue >= 100000:
    print("High Value Customer")
elif revenue >= 50000:
    print("Medium Value Customer")
else:
    print("Low Value Customer")
```

# The pass Statement

Sometimes you want to leave a block empty temporarily - Python allows that with pass.

```python
if revenue > 10000:
    pass  # Placeholder for logic to be added later
```

### Mini Dataset Example (Connecting to Data)

```python
sales_data = [
    {"region": "US", "revenue": 12000, "cost": 8000},
    {"region": "EU", "revenue": 6000, "cost": 7000},
    {"region": "APAC", "revenue": 10000, "cost": 10000},
]

for record in sales_data:
    if record["revenue"] > record["cost"]:
        print(record)
        print(record["region"], "→ Profit")
    elif record["revenue"] == record["cost"]:
        print(record)
        print(record["region"], "→ Break-even")
    else:
        print(record)
        print(record["region"], "→ Loss")
```

# In Data Engineering Context

Control flow drives ETL decisions - what to ingest, transform, or alert.
It's also used in data validation, pipeline branching, and conditional transformations.
