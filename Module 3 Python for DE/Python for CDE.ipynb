```python
salary = 95000

if salary > 100000:
    risk = "High"
elif 50000 <= salary <= 100000:
    risk = "Medium"
else:
    risk = "Low"

print(f"Salary: ₹{salary} → Risk Category: {risk}")
```

### 1.3 Loops (For & While)

```python
# Using a for loop
for i in range(5):
    print(f"Processing record {i}")

# Using a while loop
count = 0
while count < 3:
    print(f"Batch {count} processing")
    count += 1
```

```python
salaries = [40000, 52000, 61000, -1, 72000]  # -1 is a bad data value

for s in salaries:
    if s < 0:
        print("Anomaly detected! Salary cannot be negative.")
        break
    print(f"Valid salary: ₹{s}")
```

```python
values = [10, 20, 0, 40]

for val in values:
    if val == 0:
        continue
    print(f"100 / {val} = {100 / val}")
```

## Comparison Operator - Functions

```python
print(a > b)   # True
print(a == 10) # True
print(a != b)  # True
print(a <= 10) # True
```

```python
def clean_name(name):
    name = name.strip()         # Remove leading/trailing spaces
    name = name.title()         # Capitalize each word
    return name

raw_name = "   jOHn doE   "
cleaned = clean_name(raw_name)
print(f"Original: '{raw_name}' → Cleaned: '{cleaned}'")
```

```python
#Default parameter of function
def apply_tax(price, rate=0.18):
    taxed = price + (price * rate)
    return round(taxed, 2)

print(apply_tax(100))        # Uses default 18% tax
print(apply_tax(100, 0.25))  # Uses custom 25% tax
```

## Lambda and Map

```python
names = [" alice ", "BOB", "   eve "]

cleaned_names = list(map(lambda x: x.strip().title(), names))
print(cleaned_names)
```

##Filter with Lambda to remove invalid records

```python
transactions = [120, 0, -10, 450, 300, -5]

valid_transactions = list(filter(lambda x: x > 0, transactions))
print(valid_transactions)
```

## Try Accept

```python
try:
    age = int(input("Enter your age: "))
    if age < 0:
        raise ValueError("Age cannot be negative")
    print(f"Your age is {age}")
except ValueError as e:
    print(f"Invalid input: {e}")
```

##  Reduce for Aggregate

```python
from functools import reduce

amounts = [100, 200, 300]
total = reduce(lambda x, y: x + y, amounts)
print(f"Total Amount: ₹{total}")
```

### 3. Using Generators for Memory Efficiency
       - Generators allow iteration over large datasets without loading everything into memory.

```python
def large_data_generator():
    for i in range(1, 1000000):
        yield i * 2  # Generate numbers without storing them all in memory

# Accessing generator values
gen = large_data_generator()
print(next(gen))  # First value
print(next(gen))  # Second value
```

## Writing Modular Reusable Code

```python
def extract_data(file_path):
    return spark.read.option("Header","true").csv(file_path)

def transform_data(df):
    df.groupby("city").count()
    return df

def load_data(df, output_path):
    df.to_csv(output_path, index=False)

# ETL Pipeline Execution
df = extract_data("/Volumes/thebricklearning/aianalyticsengineer/datasets/csvsamples/sample_customer_data.csv")
df = transform_data(df)
display(df)
```
