### What Is pandas?

Pandas is a Python library for working with structured data - rows and columns - just like Excel or SQL tables.

- Series → one-dimensional data (like a single column)
- DataFrame → two-dimensional data (rows + columns)

### Creating Series - The Building Block

A Series represents one column of data.

```python
import pandas as pd

# Example 1: from list
s = pd.Series([100, 200, 300, 400,500], name="Salary")
print(s)
```

### Creating a DataFrame from Dictionary

A DataFrame is made up of multiple Series (columns) combined.

```python
data = { 
        "emp_id":[1,2,3,4],
        "emp_name":["John","Mary","Mike","Mary2"],
        "salary":[10000,20000,30000, 50000]
        }

employees_df = pd.DataFrame(data)
print(employees_df)
```

### Real-World Mini Exercise
Let's simulate a full pipeline with the sample tables:

```python
import pandas as pd

# Step 1: Read
employees_df = pd.read_csv("/Volumes/thedataengineering_00/data/sampledata/sampleimages/employee_panda.csv")
department_df = pd.read_csv("/Volumes/thedataengineering_00/data/sampledata/sampleimages/department_panda.csv")

# Step 2: Clean
employees_df["emp_name"] = employees_df["emp_name"].str.strip().str.title()

# Step 3: Join
final_df = employees_df.merge(department_df, on="department_id", how="left")

# Step 4: Write
final_df.to_csv("employee_department_clean.csv", index=False)

print(" Employee-Department table created successfully!")
```
