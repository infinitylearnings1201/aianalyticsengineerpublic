
#  Advanced Functions: Lambda, Map, Filter, and Modular Programming

In real-world pipelines, you often apply small transformations repeatedly - clean text, calculate ratios, validate fields, or filter rows.

Instead of writing long loops, Python offers functional shortcuts:

- lambda → write small, anonymous functions
- map() → apply a function to every element
- filter() → keep only elements that meet a condition

These are the hidden workhorses in many pandas, PySpark, and ETL transformations.

# What Is a Lambda Function?

A lambda function is a one-line, anonymous function.
 It's used when you don't want to formally define a full function with def.


          lambda arguments: expression

```python
square = lambda x: x*x

print(square(6))
```

##  B. map() - Apply Function to Every Element
map(function, iterable) → applies a function to each item and returns an iterator.

```python
#Apply discounts to all the sales values

sales = [1000,30000,5000,8000]

discount = lambda x: x*0.8

final = list(map(discount,sales))
print(final)
```

### C. filter() - Select Items Based on a Condition

filter(function, iterable) → returns only items where the function returns True.

```python
discounted_price = list(filter(lambda x:x*0.8,sales))
print(discounted_price)
```

# Combining map() and filter()

You can combine both for transformations + filtering.

```python
sales = [12000, 8000, 15000, 6000]

# Step 1: Keep high sales
filtered = filter(lambda x: x > 10000, sales)

# Step 2: Apply a 10% discount
discounted = map(lambda x: x * 0.9, filtered)

print(list(discounted))
```

# E. reduce() for Aggregation

If you need to aggregate all elements into a single result, use reduce() from functools.

```python
from functools import reduce

sales = [100, 200, 300, 400]
total = reduce(lambda x, y: x * y, sales)
print("Total Sales:", total)
```

###  DE Pipeline Example 1: Clean and Validate Data

```python
customers = [" akash ", "  Priya", "Rohan  ", ""]

# Step 1: Strip extra spaces
cleaned = list(map(lambda x: x.strip(), customers))

# Step 2: Filter out empty names
valid_customers = list(filter(lambda x: x != "", cleaned))

print(valid_customers)
```

###  DE Pipeline Example 2: Clean and Validate Data

```python
sales = [8000, 15000, 9500, 12000]

status = list(map(lambda x: "High" if x > 10000 else "Low", sales))
print(status)
```
