# Date & Time Handling - datetime, Formatting, and Timezones

Every data engineer eventually deals with timestamps - when a transaction happened, when a job ran, or how old a record is.

- Poor handling of timezones or formats leads to:
- Wrong aggregations ("sales by day" off by a few hours)
- Wrong freshness calculations
- Duplicates or missed events

That's why mastering the Python datetime module - and pandas' time utilities - is essential for reliable pipelines.

### Importing Datetime libraries

```python
from datetime import datetime, date, timedelta
import pandas as pd
import pytz  # for timezone handling
```

```python
today = date.today()
now = datetime.now()

print("Today's Date:", today)
print("Current DateTime:", now)
```

### Creating Specific Dates and Times

```python
custom_date = datetime(2025, 5, 10, 14, 30, 0)
print(custom_date)
```

### Formatting Dates (String ↔ Datetime)

Real-world data often stores dates as strings.
 We must convert and format them properly.

```python
date_str = "2025-10-07 21:30:00"
dt_obj = datetime.strptime(date_str, "%Y-%m-%d %H:%M:%S")

print(dt_obj)
print(dt_obj.strftime("%d-%b-%Y %I:%M %p"))
```

###  Date Arithmetic with timedelta
To compute differences or add/subtract days:

```python
yesterday = date.today() - timedelta(days=1)
future_date = datetime.now() + timedelta(weeks=2)

print("Yesterday:", yesterday)
print("Two weeks later:", future_date)
```

### Difference Between Two Dates

```python
start = datetime(2025, 10, 1, 10, 0)
end = datetime(2025, 10, 7, 22, 0)

diff = end - start
print(diff)
print(diff.days, "days")
print(diff.total_seconds() / 3600, "hours")
```

### Handling Timezones (Critical in Global Pipelines)

### a) Make a datetime "timezone aware"

```python
import pytz

utc = pytz.UTC
local_dt = datetime.now(utc)
print("UTC Time:", local_dt)
```

Always store timestamps in UTC in your data lake - and convert to local only for display.

### b) Convert to Another Timezone

```python
ist = pytz.timezone("Asia/Kolkata")
ist_time = local_dt.astimezone(ist)
print("IST Time:", ist_time)
```

### Handling Dates in pandas
You'll often find timestamps as columns in CSVs or APIs.

```python
data = {"event_id": [1, 2, 3],
         "event_time": ["2025-10-07 21:00:00", "2025-10-07 22:30:00", "2025-10-08 00:15:00"]}

df = pd.DataFrame(data)
df["event_time"] = pd.to_datetime(df["event_time"])
print(df.dtypes)
```

```python
df["date"] = df["event_time"].dt.date
df["hour"] = df["event_time"].dt.hour
df["weekday"] = df["event_time"].dt.day_name()
print(df)
```

### Filtering by Date Ranges

```python
filtered = df[df["event_time"] >= "2025-10-07 22:00:00"]
print(filtered)
```

### Real-World Example - Data Freshness Calculation

Let's say you ingest weather data via API.
 You want to check if the data is fresh (within last 1 hour).

```python
from datetime import datetime, timezone

last_update = datetime.fromisoformat("2025-10-07T21:30:00+00:00")
now = datetime.now(timezone.utc)

diff = (now - last_update).total_seconds() / 60
if diff < 60:
    print(" Data is fresh (updated within last hour).")
else:
    print(" Data is stale - last updated", round(diff, 1), "minutes ago.")
```
