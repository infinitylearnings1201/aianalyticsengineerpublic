# Strings in Python - Cleaning, Formatting, and Preparing Data for Processing

When you load data from files or APIs, almost everything arrives as a string - even numbers, timestamps, and identifiers.
 Before you can join, aggregate, or transform, you must clean and standardize text data.

For example:
- "  USA  " → "USA"
- "John, Doe" → "John Doe"
- "2025/10/07" → 2025-10-07

This is where Python's string operations shine.

# 1. What Is a String?

A string is a sequence of characters enclosed in quotes.

```python
name = "John1"
greeting = 'Welcome to Databricks'
multiline = """This is
a multi-line
string."""

print(name)
print(greeting)
print(multiline)
```

### Accessing and Slicing Strings

Strings behave like lists of characters - you can access or slice parts of them.

```python
word = "Databricks"

print(word[0])      # D
print(word[-1])     # s
print(word[0:4])    # Data
print(word[4:])     # bricks
```

[:] syntax helps extract substrings - handy for column parsing (like splitting "2025-10-07" into "2025").

### String Immutability (But Reassignment possible)
Strings cannot be changed once created.

```python
text = "data"

text[0] = 'D' # Error
```

```python
text = "data"
# text[0] = 'D' Error
text = "Data"   #Reassignment
```

###  This is important when you clean data - you're always creating new strings, not editing in place.

### Common String Methods for Data Cleaning

###  Real-World Cleaning Example

```python
names = ["  Akash ", "PRIYA ", " rOhan  ", "", "jOhn     "]

cleaned = []

for n in names:
    n = n.strip().title()   # Remove spaces, capitalize
    if n != "":
        cleaned.append(n)

print(cleaned)
```

### Checking and Validating Strings

```python
email = "user@example.com"
phone = "9876543210"

print(email.endswith(".com"))   # True
print(phone.isdigit())          # True
print(email.isalpha())          # False

if (phone.isdigit() and len(phone) == 10):
    print("Valid phone number")
else:
    print("Invalid phone number")
```

# Removing Special Characters

Sometimes data comes with unwanted punctuation, symbols, or hidden characters.

```python
import re

text = "Hello@@ World##!!"
cleaned = re.sub(r'[^A-Za-z0-9 ]+', '', text)

print(text)
print(cleaned)
```

### Formatting Strings (f-strings)

f-strings allow you to embed variables inside text - great for logs and ETL messages.

```python
table = "sales"
records = 2500
print(f"Loaded {records} records into {table} table.")
```

### Combining String Operations

Realistic scenario: cleaning file names before ingestion

```python
files = [" Sales_Jan.CSV ", "sales_Feb.csv", "temp.txt"]

for f in files:
    cleaned = f.strip().lower().replace(".csv", "")
    if "sales" in cleaned:
        print("Valid file:", cleaned)
    else:
        print(" Skipped:", cleaned)
```

### Multi-line Cleaning Example - Combining All Concepts

```python
raw_data = [
    "   INFOSYS pvt ltd  ",
    "InfoSys   LTD",
    "infosys private limited",
    "   ",
]

cleaned_data = [
    re.sub(r'\s+', ' ', x.strip().title())
    for x in raw_data
    if x.strip() != ""
]

print(cleaned_data)
```
