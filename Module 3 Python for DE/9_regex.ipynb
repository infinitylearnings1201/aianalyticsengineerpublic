# Data from the real world is messy - and most of that mess is in text.

 Think about examples you'll face every day:
- Removing unwanted characters from customer names
- Validating email or phone formats
- Extracting numbers or dates from log strings
- Detecting IDs like INV-1234 or TXN_2025_01

Regex (Regular Expressions) lets you describe a text pattern and find or fix those patterns instantly - like a text search engine for your data.

### What Is Regex?

Regex stands for Regular Expression - a mini language for describing search patterns in text.
You use it through Python's built-in re module.

- re.findall()
- re.match()
- re.sub()
- re.split()

```python
import re
```

### 1. Basic Regex Functions

```python
import re

text = "Data engineers love data and big data platforms."
matches = re.findall("data", text, flags=re.IGNORECASE)
print(matches)
```

### 3. Extracting Numbers from Text

```python
import re
text = "Order ID: 4532, Amount: 679089 USD"
numbers = re.findall(r'\d+', text)
print(numbers)
```

### 4. Validating Email Addresses

```python
import re

emails = ["user@example.com", "info@data", "test@domain.co.in"]

pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'

for e in emails:
    if re.match(pattern, e):
        print(" Valid:", e)
    else:
        print(" Invalid:", e)
```

### 5. Extracting Dates from Log Messages

```python
log = "2025-10-08 Process started at 09:00; 2025-10-08 completed."
dates = re.findall(r'\d{4}-\d{2}-\d{2}', log)
print(dates)
```

### 6. Cleaning Special Characters from Text

```python
text = "Hello@@ World##!! Data-Engineering$$ Rocks"
cleaned = re.sub(r'[^A-Za-z0-9 ]+', '', text)
print(cleaned)
```

You just removed everything except alphabets, numbers, and spaces - a typical step before writing clean data downstream.

### 7. Masking Sensitive Information (PII Example)
Let's hide customer emails before saving logs

```python
data = "Customer email address: john.doe@example.com"
masked = re.sub(r'[\w\.-]+@[\w\.-]+\.\w+', '[EMAIL MASKED]', data)
print(masked)
```

Regex is powerful for data governance and PII masking in audit pipelines.

### 8. Split Strings with Multiple Delimiters

Instead of just split(','), regex can handle multiple separators.

```python
text = "data,engineering;python|spark"
parts = re.split(r'[;,|]', text)
print(parts)
```

### 9. Real-World Scenario - Validate Customer Records

```python
records = [
    {"email": "john@info.com", "phone": "9876543210"},
    {"email": "akash@", "phone": "9988"},
    {"email": "priya@data.org", "phone": "9876543212"},
]

email_pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
phone_pattern = r'^\d{10}$'

for r in records:
    valid_email = bool(re.match(email_pattern, r["email"]))
    valid_phone = bool(re.match(phone_pattern, r["phone"]))

    if valid_email and valid_phone:
        print(" Valid record:", r)
    else:
        print(" Invalid record:", r)
```
