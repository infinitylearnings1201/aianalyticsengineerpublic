# Python Pandas Example

```python
import pandas as pd

# small dataset
data = {
    "region": ["North", "South", "East", "West"],
    "revenue": [1000, 1500, 1200, 1700]
}

df_pandas = pd.DataFrame(data)

# view data
print(df_pandas)

# calculate average revenue
print("Average Revenue:", df_pandas["revenue"].mean())
```

## Pandas has scalability issues.

# Spark DataFrames: Built for Scale

```python
from pyspark.sql import SparkSession

# start Spark
spark = SparkSession.builder.appName("day3-demo").getOrCreate()

# same dataset
data = [("North", 1000), ("South", 1500), ("East", 1200), ("West", 1700)]
columns = ["region", "revenue"]

df_spark = spark.createDataFrame(data, columns)

# view data
df_spark.show()

# calculate average revenue
from pyspark.sql.functions import avg
df_spark.select(avg("revenue")).show()
```

Looks very similar, right? But here's the difference: Spark doesn't care if it's 4 rows or 4 billion. The data is distributed across machines, and Spark executes your query in parallel.

As freshers, you may love the simplicity of pandas. But remember, in data engineering we rarely deal with small toy datasets. The real world is messy and massive.
A retail company may process billions of sales transactions every day.
A telecom company ingests millions of events per second from devices.
A bank may analyze decades of financial records for compliance.

# Hybrid Use: Pandas + Spark

```python
# Spark DataFrame
result = df_spark.groupBy("region").sum("revenue")

# convert to pandas
pdf = result.toPandas()
print(pdf)
```

Here's the takeaway from today:

Pandas is great for small to medium data on a single machine.

Spark DataFrames are built for distributed, large-scale workloads.

Syntax is similar, so your pandas knowledge transfers easily to Spark.

In practice, data engineers often use Spark + pandas together for different stages.
