Whitepaper: When to Use What Type of Ingestion on Databricks Lakeflow
Executive Summary

Modern data platforms operate in an environment where information arrives at different speeds, volumes, structures, and levels of reliability. Whether your data originates from enterprise systems, cloud object storage, IoT devices, SaaS applications, or operational databases, your ingestion strategy determines the efficiency, cost, reliability, and scalability of your entire Lakehouse.

Databricks Lakeflow unifies ingestion through Connect, Declarative Pipelines, and Jobs, providing a single platform to bring raw data into Delta Lake—no matter the complexity of sources. This whitepaper explains when to use which ingestion pattern within Lakeflow’s ecosystem, enabling organizations to design ingestion frameworks that are resilient, scalable, and business-aligned.

Introduction: Why Ingestion Strategy Matters

Data ingestion is no longer a simple ETL step. It is the front door to the Lakehouse—the first point of contact where quality, lineage, and trust are established.

A modern ingestion layer must answer questions like:

Do we need the freshest possible data or periodic snapshots?

Should ingestion be incremental, continuous, or triggered?

How do we track metadata and handle malformed records?

What about schema drift, late-arriving data, and CDC?

How do we ingest from databases, SaaS APIs, and cloud storage at scale?

Lakeflow simplifies this complexity by offering ingestion methods that serve different operational goals—from one-time loads to high-throughput streaming.

This document explains each method and provides guidance on choosing the right technique based on the characteristics of your workloads.

The Four Pillars of Ingestion in Databricks Lakeflow

Although Lakeflow offers many capabilities, ingestion patterns fall under four architectural pillars:

Batch ingestion (CTAS)

Incremental ingestion (COPY INTO, Auto Loader incremental mode)

Continuous ingestion (Auto Loader streaming, event streams)

Enterprise ingestion (Managed Connectors, CDC, MERGE INTO)

Each pillar aligns with a different operational scenario.

Let’s examine their purpose, strengths, and ideal use cases.

1. Batch Ingestion: When You Need Simple, One-Time, or Full Loads

Batch ingestion remains relevant because many workflows still expose static or infrequently changing data. Using CREATE TABLE AS SELECT (CTAS) and simple spark.read operations, batch ingestion offers the fastest path from raw files to Delta.

Batch ingestion is best when:

The entire dataset needs to be reloaded or backfilled.

Data volume is manageable (hundreds of GB, not petabytes).

You’re building or refreshing a Bronze table for the first time.

Data updates are rare, or downstream consumers expect full snapshots.

A typical CTAS workload could be loading historical sales data or migrating a table from an existing warehouse into Delta.

Batch ingestion trades freshness for simplicity. It reloads everything—so it should be used where idempotency and incremental cost are not critical concerns.

2. Incremental Batch Ingestion: When Files Arrive Periodically

Many operational systems drop new files into cloud storage every hour, every day, or based on business events. These datasets tend to be append-only, and the goal is to process only the new files without scanning the entire directory.

This is where COPY INTO and Auto Loader’s incremental trigger shine.

These methods are ideal when:

New files regularly appear in storage locations.

You want ingestion to be lightweight and cost-effective.

There is no need for real-time processing.

The source is append-only or quasi-append-only.

Schema may evolve, but in a controlled manner.

COPY INTO is simple SQL, idempotent, and reliable. It is perfectly suited for pipelines like:

Daily orders

Incremental billing files

Hourly application logs

IoT device batches uploaded every few minutes

Incremental batch ingestion avoids unnecessary reprocessing and is optimized for ongoing ingestion from object storage.

3. Continuous Streaming Ingestion: When You Need Low Latency and High Throughput

Modern analytics depends increasingly on streaming data—clickstreams, IoT sensors, payments, telemetry, and events. These workloads require ingestion frameworks that adapt to unpredictable arrival patterns, high throughput, and frequent schema changes.

Databricks’ Auto Loader with continuous mode, combined with Structured Streaming, allows pipelines to ingest billions of files with minimal operational overhead.

Streaming ingestion is the right choice when:

Data must be available in near real-time.

You expect irregular arrival patterns.

High throughput is required and files may arrive out of sequence.

You want the system to detect and evolve schemas automatically.

You require streaming Bronze → streaming Silver → Gold transformations.

Auto Loader provides:

Event-driven ingestion

Automatic schema evolution

Fault tolerance through checkpoints

Ability to handle massive file counts

If your workload has “always-on” data—like real-time logistics, machine monitoring, fraud detection, or digital engagement—streaming ingestion is essential.

4. Enterprise Ingestion: When You Integrate Databases, SaaS Apps, or CDC Streams

Organizations rarely operate on files alone. Data often lives in:

Oracle, SQL Server, Postgres, MySQL

SaaS platforms like Salesforce, ServiceNow, Workday

Event buses & message queues like Kafka

Operational systems producing log-based CDC streams

This is where Lakeflow Connect Managed Connectors and CDC with MERGE INTO become critical.

Enterprise ingestion is recommended when:

The data source is operational (OLTP), not file-based.

You need continuous synchronization between the system of record and the Lakehouse.

You want zero-code connectors with serverless compute.

You want change data capture (CDC) applied to Delta tables.

You need governance and lineage enforced through Unity Catalog.

Managed Connectors provide:

Serverless ingestion jobs

Automatic incremental pulls

Secure credential management

Prebuilt integrations with popular enterprise systems

Once data lands in Bronze, MERGE INTO is the tool of choice to maintain accurate Silver tables:

Handling upserts

SCD Type 1 or Type 2 logic

Deduplication

Late-arriving updates

Business-driven state changes

Enterprise ingestion keeps your Lakehouse in sync with business systems in a continuous and governed manner.

Choosing the Right Method: A Decision Framework

Rather than memorizing definitions, it’s better to evaluate ingestion patterns based on five decision vectors:

1. Freshness Requirements

Needs real-time? → Streaming (Auto Loader) or Kafka

Needs periodic updates? → Incremental (COPY INTO)

Needs only initial load or snapshots? → Batch (CTAS)

2. Source Type

Files (CSV/JSON/Parquet)? → CTAS / COPY INTO / Auto Loader

Database? → Managed Connector + CDC

SaaS Application? → Managed Connector

Event stream? → Kafka / Auto Loader streaming

3. Data Volume & Frequency

High volume / high frequency → Streaming ingestion

Medium volume / scheduled → Incremental batch

Stable or static → Batch ingestion

4. Complexity of Updates

Append-only → COPY INTO or Auto Loader

Change events (updates/deletes) → MERGE INTO + CDC

5. Governance & Observability Needs

Enterprise-grade security → Managed Connectors + UC

Simple raw ingestion → CTAS / COPY INTO

High traceability → Metadata columns + Delta logs

Putting It All Together: Architectural Recommendations
Use CTAS when

You need to ingest an entire dataset at once, usually for initial load or backfills, and the cost of reprocessing is acceptable.

Use COPY INTO when

Your source produces new files regularly, and ingestion must be incremental, idempotent, and reliable with minimal orchestration.

Use Auto Loader streaming when

Latency matters, files arrive continuously, and schema evolution has to be automatic, especially for massive or unpredictable workloads.

Use Managed Connectors when

You ingest data from enterprise systems, need secure credential governance, want serverless ingestion, and prefer a no-code or low-code ingestion experience.

Use MERGE INTO when

You must apply upserts, deletes, deduplication, or SCD logic to maintain accurate downstream tables in Silver and Gold layers.
